a) Look for the missing values in all the columns and either impute them (replace with mean,
median, or mode) or drop them. Justify your action for this task.

```python
missing_data = trainn.isnull().sum()
print(missing_data) # Checking for missing values

miss_col = []
for col, value in missing_data.items():
    if value > 0:
        miss_col.append(col)

print("Columns with missing values:", miss_col)
```

```python
for col in miss_col:
    if trainn[col].dtype == 'object':
        trainn[col].fillna(trainn[col].mode()[0], inplace=True)
    else:
        trainn[col].fillna(trainn[col].mean(), inplace=True)
```

C) Change the categorical variables (“Fuel_Type” and “Transmission”) into numerical one hot
encoded value.

```python
labelling = LabelEncoder()
trainn['Fuel_Type_Label'] = labelling.fit_transform(trainn['Fuel_Type'])
trainn['Transmission_Label'] = labelling.fit_transform(trainn['Transmission'])

onehot_encoder = OneHotEncoder(sparse=False)

fuel_type = onehot_encoder.fit_transform(trainn[['Fuel_Type_Label']])
transmission = onehot_encoder.fit_transform(trainn[['Transmission_Label']])

fuel_type_df = pd.DataFrame(fuel_type, columns=['Fuel_Type_' + str(i) for i in range(fuel_type.shape[1])])
transmission_df = pd.DataFrame(transmission, columns=['Transmission_' + str(i) for i in range(transmission.shape[1])])

trainn = pd.concat([trainn, fuel_type_df, transmission_df], axis=1)

trainn.drop(['Fuel_Type', 'Transmission', 'Fuel_Type_Label', 'Transmission_Label'], axis=1, inplace=True)

```
